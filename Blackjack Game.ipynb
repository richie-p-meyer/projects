```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

class Card:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        return f'{self.rank} of {self.suit}'

class Deck:
    def __init__(self):
        self.allcards = []
        for s in suits:
            for r in ranks:
                self.allcards.append(Card(s,r))
    
    def shuffle(self):
        random.shuffle(self.allcards)
        
    def deal_one(self):
        return self.allcards.pop(0)

class Player:
    def __init__(self):
        self.name = input(f'What is your name? ')
        self.money = int(input(f'{self.name}, how much money are you starting with? '))
        self.players_cards = []
    
    def __str__(self):
        return f'Hello {self.name}, you have ${self.money} - May the odds be ever in your favor!'
    
    def hit(self,cards):
        return self.players_cards.append(cards)
    
    def win_money(self, wins):
        self.money += wins
    
    def lose_money(self, losses):
        self.money -= losses
        
    def cards_value(self):
        total = 0
        for f in range(0,len(self.players_cards)):
            total+=self.players_cards[f].value
        return total
    
    def print_hand(self):
        for r in range(0,len(self.players_cards)):
            print(self.players_cards[r])

class Dealer:
    def __init__(self):
        self.dealers_cards = []
         
    def deal_dealer(self, deck):
        self.dealers_cards.append(deck.deal_one())
        self.dealers_cards.append(deck.deal_one())
    
    def show_dealer_card(self):
        print(self.dealers_cards[0])
    
    def cards_value(self):
        total = 0
        for f in range(0,len(self.dealers_cards)):
            total+=self.dealers_cards[f].value
        return total
    

    
    

    
    
    
    
```

```python
#Game Setup
deck = Deck()
deck.shuffle()

#Determine how many players and create players

player_count = int(input("How many players will be playing? "))
players = list()
bet = list()

for p in range(0,player_count):
    print(f'Player {p+1}')
    players.append(Player())


print('Welcome to the table, may the odds be ever in your favor')


game_on = True
while game_on:
    #Betting phase
    bet = []
    for p in range(0, player_count):
        bet.append(int(input (f'{players[p].name} how much would you like to bet? \n')))
        while bet[p]>players[p].money:
            bet[p] = (int(input (f'{players[p].name} how much would you like to bet? \n')))

    #Deal 2 Cards
    for p in range(0, player_count):
        players[p].players_cards.append(deck.deal_one())
        players[p].players_cards.append(deck.deal_one())

    #Show Players their cards
    for p in range(0, player_count):
        print(f'\n{players[p].name}\n{players[p].players_cards[0]}\n{players[p].players_cards[1]}\n{players[p].cards_value()}')
        if players[p].cards_value() == 21:
            print(f'Blackjack for {players[p].name}!')
            print(f'{players[p].name} won ${bet[p]*1.5}')
            players[p].win_money(bet[p]*1.5)


    #Setup Dealer
    dealer = Dealer()
    dealer.deal_dealer(deck)
    print(f'\nDealer is showing:')
    dealer.show_dealer_card()
    print('\n')


    while dealer.cards_value() < 17:
        dealer.dealers_cards.append(deck.deal_one())



    #Ask Players to hit or to stay

    for p in range(0, player_count):
        hit = 'Hit'

        while hit == 'Hit':

            if input(f'\n{players[p].name}: Hit or Stay? ').capitalize() == 'Hit':
                players[p].hit(deck.deal_one())
                players[p].print_hand()
                print(players[p].cards_value())
                hand_value = players[p].cards_value()
                if hand_value > 21:
                    print(f'{players[p].name} has BUSTED - Loser! ')
                    break
                else:
                    pass

            else:
                break

    print("\nDealer's Hand: ")
    for p in range(0, len(dealer.dealers_cards)):
        print(dealer.dealers_cards[p])
    if dealer.cards_value() > 21:
        print('Dealer has BUSTED!')
    else:
        print(dealer.cards_value())
    print("\n")

    for p in range(0, player_count):
        if players[p].cards_value() == 21 and len(players[p].players_cards) == 2:
            pass     
        
        elif players[p].cards_value() < 22 and players[p].cards_value() > dealer.cards_value():
            players[p].players_cards = []
            players[p].win_money(bet[p])  
            print(f'{players[p].name} has won ${bet[p]}')

        elif players[p].cards_value() < 22 and dealer.cards_value() > 21:
            players[p].players_cards = []
            players[p].win_money(bet[p])  
            print(f'{players[p].name} has won ${bet[p]}')

        elif players[p].cards_value() < 22 and players[p].cards_value() == dealer.cards_value():
            players[p].players_cards = []
            print(f'{players[p].name} has pushed')

        else:
            players[p].lose_money(bet[p]) 
            players[p].players_cards = []
            print(f'{players[p].name} has lost ${bet[p]}')
            
 
    another_hand = input('Would you like to play another hand?').capitalize()
    
    if another_hand != 'Yes':
        break
    



```
