```python
from datetime import datetime

class Bank_Account:
    def __init__(self):
        self.money = round(float(input(f'How much is your initial deposit? ')),2)
        self.transaction_history = []
        print(f'\nGreat! Your account now has ${self.money}')

    def __str__(self):
        return f'This account has ${self.money}.'
    
    def withdraw(self):
        withdrawal = float(input("How much money to withdraw?\n"))
        
        if self.money >= withdrawal:
            self.money -= withdrawal
            
            now = datetime.now()
            time = now.strftime("%H:%M:%S")
        
            post_description = input('Would you like to add a note? "Y" or "N"\n').lower()
            if post_description == 'y':
                description = input('\nPlease enter note\n')
                self.transaction_history.append(f'{time} Withdrawal ${withdrawal} {description}')
            else:
                self.transaction_history.append(f'{time} Withdrawal ${withdrawal}')
           
            print(f'You now have ${self.money}')
            
            
            
        else:
            print(f'Not enough funds. You have ${self.money}')
               
    def deposit(self):
        deposit_money = float(input("How much money to deposit?\n"))
        self.money += deposit_money
        
        now = datetime.now()
        time = now.strftime("%H:%M:%S")
        
        post_description = input('Would you like to add a note? "Y" or "N"\n').lower()
        if post_description == 'y':
            description = input('\nPlease enter note\n')
            self.transaction_history.append(f'{time} Deposit ${deposit_money} {description}')
        else:
            self.transaction_history.append(f'{time} Deposit ${deposit_money}')
        
        
        print(f'You now have ${self.money}')
        
    
    
    def view_balance(self):
        print(f'You have ${self.money}')
        
              
    
    def view_transactions(self):
        for t in range(len(self.transaction_history)):
            print(self.transaction_history[t])
        
```

```python
print('~~~ Welcome to your terminal checkbook! ~~~\n')

account=Bank_Account()

while True:
    print('\nWhat would you like to do?\n')
    print('1: View Current Balance')
    print('2: Make a Withdrawal')
    print('3: Make a Deposit')
    print('4: View Transactions')
    print('5: Exit')
    choice = input('')
    
    if choice == '1':
        account.view_balance()
        
    elif choice == '2':
        account.withdraw()
        
    elif choice == '3':
        account.deposit()
        
    elif choice == '4':
        account.view_transactions()
        
    else:
        break
    

```
